```python
Liga= df.loc[df.Tournament=='LaLiga']
```

```python
Liga.head()
```

```python
Liga[['Team','Goals']].sort_values(by='Goals',ascending=False)[:5].plot(kind='bar',x='Team',y="Goals",cmap='Set2')
plt.title("Goles totales de los Mejores 5 Equipos de LaLiga",fontsize=15)
plt.show()
```

```python
sns.jointplot(x='Shots pg',y='Team',data=Liga)
plt.title("Tiros por Juego vs Equipo en LaLiga",fontsize=15,loc="right")
plt.show()
```

```python
top_shots_pg=Liga[['Team','Goals','Shots pg']]
```

```python
top_shots_pg.head()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='Team',y='Goals',data=top_shots_pg[:5])
plt.title("Line Plot de Goles vs Equipo en LaLiga",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(7,7))
labels = top_shots_pg.index
plt.pie(x='Shots pg',data=top_shots_pg,labels='Team',startangle=90, autopct='%.1f%%')
plt.title("Tiros por Juego de cada equipo en LaLiga \n 2020/21 Temporada Completa", fontsize = 24) 

plt.tight_layout() 
plt.show()
```

```python
LaLiga_possesion=Liga.sort_values(by='Possession%',ascending=False)
```

```python
LaLiga_possesion.head()
```

```python
sns.barplot(x='Team',y='Possession%',data=LaLiga_possesion[:5],palette='Paired')
plt.title("Posesion Maxima de LaLiga", fontsize=15)
plt.show()
```

```python
sns.catplot(x='Goals',y='Shots pg',data=Liga[:5],hue="Team")
plt.title("Goles vs Tiros por Partido (Mejores 5)",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,10))
tournaments=df.groupby('Tournament')['Goals','Pass%','Possession%','Shots pg'].sum().plot(kind='bar')
plt.title(" Comparacion de los torneos", fontsize=15)
plt.show()
```
